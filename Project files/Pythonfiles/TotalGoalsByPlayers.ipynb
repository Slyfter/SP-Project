```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate

df = pd.read_csv("data\data.csv", encoding = "ISO-8859-1", sep=";")

goals=list(round(df['Goals']*df['90s']))
shots=list(df['Shots']*df['90s'])
players=list(df['Player'])

plt.figure(figsize=(12,8))
plt.scatter(goals,shots,s=50,alpha=0.5)
plt.axhline(y=80, color='red')
plt.axvline(x=17.5, color='red')
plt.xlabel("Goals")
plt.ylabel("Shots")
for i in range(len(players)):
    if(goals[i]> 17 and shots[i]> 60):
        plt.annotate(players[i],(goals[i], shots[i]), ha='left')
```

```python
new = df[['Player', 'Pos', 'Squad', 'Comp']].copy()

new['Goals'] = round(df['Goals'] * df['90s'])

newsorted = new.sort_values('Goals', ascending=False)

newsortedtop10 = newsorted.head(10)

newsortedtop10grouped = newsortedtop10.groupby('Comp').count()

newsortedtop10grouped['Players in top 10'] = newsortedtop10grouped['Player']

newsortedtop10grouped = newsortedtop10grouped.drop(columns=['Player','Pos','Squad','Goals'])

newsortedtop10grouped

print(tabulate(newsortedtop10grouped.sort_values('Players in top 10', ascending=False), headers="keys"))
```
